```python
# =====================================
# MINI PROJECT
# Student Marks Analysis
# Using NumPy, Pandas, Matplotlib
# =====================================

# STEP 1: Import Libraries

import numpy as np
import pandas as pd
import matplotlib.pyplot as plt


# STEP 2: Upload File

from google.colab import files
uploaded = files.upload()


# STEP 3: Load Dataset

df = pd.read_csv('Simple 100 Student Marks 1.csv')

# Display first 5 rows
print(df.head())


# =====================================
# PANDAS OPERATIONS
# =====================================

print("\nDataset Shape:")
print(df.shape)

print("\nColumn Names:")
print(df.columns)

print("\nStatistical Summary:")
print(df.describe())


# =====================================
# NUMPY OPERATIONS
# =====================================

maths = np.array(df['maths_marks'])

print("\nAverage Maths Marks:", np.mean(maths))
print("Highest Maths Marks:", np.max(maths))
print("Lowest Maths Marks:", np.min(maths))
print("Standard Deviation:", np.std(maths))


# =====================================
# GRAPH 1 : LINE PLOT
# =====================================

plt.figure(figsize=(10,5))

plt.plot(df['maths_marks'],marker="*",color="#7856")

plt.title(" Marks Line Plot")
plt.xlabel("Student Index")
plt.ylabel("Maths Marks")

plt.show()


# =====================================
# GRAPH 2 : BAR PLOT
# =====================================

plt.figure(figsize=(12,5))

plt.bar(df['student_name'], df['science_marks'])

plt.title("Marks Bar Plot")
plt.xlabel("Student Name")
plt.ylabel("Science Marks")

plt.xticks(rotation=90)

plt.show()


# =====================================
# EXTRA ANALYSIS
# =====================================

average_total = (
    df['maths_marks'] +
    df['science_marks'] +
    df['english_marks'] +
    df['social_studies_marks'] +
    df['Hindi_marks']
) / 5

print("\nAverage Total Marks of Students:")
print(average_total.head())


# Pass/Fail Analysis

passed_students = np.sum(average_total >= 40)
failed_students = np.sum(average_total < 40)

print("\nPassed Students:", passed_students)
print("Failed Students:", failed_students)
```
